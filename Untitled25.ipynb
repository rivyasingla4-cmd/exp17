```python
import matplotlib.pyplot
import seaborn as sns
import pandas as pd

data = {
'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
'Study_Hours': [2, 3, 2.5, 4, 3.5],
'Marks': [50, 60, 55, 70, 65],
'Attendance': [60, 70, 65, 80, 75],
'Sleep_Hours': [6, 7, 6, 8, 7],
'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
#1 line chart (trend

plt.plot(df['Days'], df['Study_Hours'],marker='o')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('rivya')
plt.show()
```

```python
#advance line graph
plt.figure(figsize=(7,4))

plt.plot(df['Days'], df['Study_Hours'],
         marker='o', color='b', label='Study Hours')

plt.plot(df['Days'], df['Marks'],
         marker='s', color='red', label='Marks')

plt.title("rivya")
plt.xlabel("Days")
plt.ylabel("values")
plt.legend()
plt.show()
```

```python
#bar chart
plt.bar(df['Days'], df['Marks'])
plt.xlabel("Days")
plt.ylabel("Marks")
plt.title("rivya")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}

df = pd.DataFrame(data)

df

# Code 6: Bar Chart Advanced
bars = plt.bar(df['Days'], df['Marks'],color='red')
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y, str(y), ha='center', va='bottom')
plt.title('Rivya')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)

#conditional scatter plot
#example category
df['Result']=['Pass','Fail','Pass','Pass','Fail']
plot_colors = ['red' if r=='Fail' else 'green' for r in df['Result']]
plt.scatter(df['Study_Hours'],df['Marks'],c=plot_colors)
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('rivya')
plt.show()
```
