```python
# Code 1: Importing data to use for plotting
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Code 2: Dataset importing
import pandas as pd

data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}

df = pd.DataFrame(data)

df
```

```python
# Code 3: Line Chart (Trend)

plt.plot(df['Days'], df['Study_Hours'], marker='*')

plt.title("Rachit")
plt.xlabel("Days")
plt.ylabel("Study Hours")

plt.show()
```

```python
# Code 4: Advanced Line Chart

plt.figure(figsize=(7, 4))

plt.plot(df['Days'], df['Study_Hours'],
         marker='o', color='blue', label='Study Hours')

plt.plot(df['Days'], df['Marks'],
         marker='s', color='red', label='Marks')

plt.title("Rachit")
plt.xlabel("Days")
plt.ylabel("Values")

plt.legend()
plt.show()
```

```python
# Code 5: Bar chart
plt.bar(df['Days'], df['Marks'],color='green')
plt.title('Rachit')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
# Code 6: Bar Chart Advanced
bars = plt.bar(df['Days'], df['Marks'],color='red')
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y, str(y), ha='center', va='bottom')
plt.title('Rachit')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y')
plt.show()
```

```python
# Code 7: Histogram for Distribution
# Note: It shows how values are spread

data1 = {
    'Marks': [50, 60, 55, 70, 65, 75, 80, 85, 90, 45, 68, 72, 77, 82, 88]
}

df = pd.DataFrame(data)

plt.hist(df['Marks'], bins=5, edgecolor='black', alpha=0.7)

plt.title("Marks Distribution - Rachit")
plt.xlabel("Marks")

plt.show()
```

```python
# Code 8: Histogram commands learning
import matplotlib.pyplot as plt
import numpy as np
# Example data import:
Marks = [50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]

mean_Value = np.mean(Marks)
plt.figure(figsize=(7,4))
plt.hist(Marks,bins=6,color='blue',linestyle='--',linewidth=2,edgecolor='black')
plt.axvline(mean_Value,color='red',linestyle='--',linewidth=2,label='Mean')
plt.title("Rachit")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y',linestyle='--',alpha=0.3)

# This code shows a dashed line through the histogram to signify the mean.
```

```python
# Code 9: Scatter Plot
import matplotlib.pyplot as plt
import pandas as pd

data2 = {
    'Days': ['Mon','Tues','Wed','Thurs','Fri'],
    'Study_hours': [2,3,2.5,4,3.5],
    'Marks': [50,60,55,70,65],
    'Attendance': [60,70,65,80,75],
    'Sleep_Hours':[6,7,6,8,7],
    'Assignment_Completed': [1,2,1,3,2]
}

df = pd.DataFrame(data2)

plt.scatter(df['Study_hours'],df['Marks'])
plt.title("Rachit")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
# Code 10: Conditional Scatter Plot
# Example category is given below.

df['Result'] = ['Fail', 'Pass', 'Fail', 'Pass', 'Pass']

colors = ['red' if r == 'Fail' else 'green' for r in df['Result']]

plt.scatter(df['Study_hours'], df['Marks'], c=colors)

plt.title("Rachit")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.show()
```

```python
# Code 11: New dataset importing
data3 = {
    'Days': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun','Mon','Tue','Wed'],
    'Region': ['North','South','East','West','North','South','East','West','North','South'],
    'Sales': [200,220,250,270,300,350,400,280,260,290],
    'Profit': [20,25,30,35,50,60,80,40,38,45],
    'Customers': [30,35,40,45,50,65,80,55,48,52],
    'Category': ['Food','Clothing','Electronics','Food','Clothing','Electronics','Food',
                 'Clothing','Electronics','Food']
}

df = pd.DataFrame(data3)
df
```
